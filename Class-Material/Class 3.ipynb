```python
data = {
    'Class': ['D', 'A', 'B', 'C', 'A', 'B', 'A', 'D'],
    'Gender' : ['F', 'M', 'T', 'M', 'F', 'T', 'M', 'M'],
    'Age' : [20, 25, 12, 12, 25, 24, 16, 20],
    'Class1': ['D', 'A', 'B', 'C', 'A', 'B', 'A', 'D'],
    'Gender1' : ['D', 'A', 'B', 'C', 'A', 'B', 'A', 'D'],
}
```

```python
df3 = pd.DataFrame(data)
```

```python
cat_cols0 = ['Class', 'Gender', 'Class1', 'Gender1']
```

```python
cat_cols0
```

```python
for col0 in cat_cols0:
    df3[col0] = df3[col0].astype('category').cat.codes
```

```python
cat_cols1 = df3.select_dtypes(include=['object']).columns
```

```python
cat_cols1
```

```python
for col1 in cat_cols1:
    df3[col1] = df3[col1].astype('category').cat.codes
```

```python
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.plot()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.title('Scatter Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker')
plt.title('Scatter Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
tips
```

```python
tips['sex'].unique()
```

```python
sns.lineplot(x='day', y='total_bill', data=tips)
plt.title('Line Plot')
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=25) #1, 2, 3, 4, 5, 4, 5, 4, 6, 7, 8, 10
plt.title('Histogram')
plt.show()
```

```python
tips['day'].unique()
```

```python
#5 no summary: minimum, Q1 -> 25%, median ~ 50%, Q3 -> 75%, maximum
#Box Plot -> Inter Quartile Range: Q3-Q1 , Q3 = 3/4*(N), Q1=1/4*(N) 
sns.boxplot(x='day', y='total_bill', data=tips) 
plt.title('Box Plot')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips) 
plt.title('Violin Plot')
plt.show()
```

```python
sns.pairplot(tips, hue='day')
```

```python
plt.bar(tips.day, tips.day.values) 
plt.title('Bar Plot')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips) #line-> confidence interval
plt.title('Bar Plot')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```
