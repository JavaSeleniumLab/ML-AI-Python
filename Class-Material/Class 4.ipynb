```python
from scipy.interpolate import griddata
x = iris['sepal_length']
y = iris['sepal_width']
z = iris['petal_length']

xi = np.linspace(x.min(), x.max())
yi = np.linspace(y.min(), y.max())
xi, yi = np.meshgrid(xi, yi)
zi = griddata((x, y), z, (xi, yi))

fig = plt.figure(figsize=(20, 15))
ax3d = fig.add_subplot(111, projection='3d')#2d -> x,y = 1,1, 3d -> x,y,x=1,1,1
ax3d.plot_surface(xi, yi, zi, cmap='RdYlGn')
ax3d.set_xlabel('Sepal Length')
ax3d.set_ylabel('Sepal Width')
ax3d.set_zlabel('Petal Length')
plt.show()
```

Dot Product

```python
a = [1, 2, 4]
b = [1, 5, 2]
#1*1 + 2*5 + 4*2
```

```python
def dot_product(x, y):
    return sum(i*j for i, j in zip(x, y))
dot_product(a, b)
```

Cross Product

```python
def cross_product(x, y):
    d1 = x[0]*y[1] - x[1]*y[0] #1*5 - 2*1 = 5-2 = 3
    d2 = x[1]*y[2] - x[2]*y[1] #2*2 - 4*5 = 4-20 = -16
    d3 = x[2]*y[0] - x[0]*y[2] #4*1 - 1*2 = 4-2 = 2
    return [d1, d2, d3]

a = [1, 2, 4]
b = [1, 5, 2]

cross_product(a, b) #
```

```python
def vectors(v):
    dot_pro = sum(i*i for i in v) #1+4+9+16+25 = 55
    return math.sqrt(dot_pro) #square_root(55)

vectors(v)
```

```python
def matrix_add(x, y):
    xrows = len(x)
    xcols = len(x[0])
    yrows = len(y)
    ycols = len(y[0])
    
    if xrows != yrows or xcols != ycols:
        print('Size not same')
    else:
        res = [[0 for i in range(xcols)] for i in range(xrows)]
        for i in range(xrows):
            for j in range(xcols):
                res[i][j] = x[i][j] + y[i][j]
        return res
```

```python
def determinant_2x2(matrix):
    if len(matrix)==2 and all(len(row)==2 for row in matrix):
        a, b = matrix[0]
        c, d = matrix[1]
        return (a*d) - (c*b)

matrix1 = [[2, 3],
          [4, 5]]

print(determinant_2x2(matrix1))
```

```python
#https://media.geeksforgeeks.org/wp-content/uploads/20200709193253/Screenshot28.png
def determinant_3x3(matrix):
    if len(matrix)==3 and all(len(row)==3 for row in matrix):
        a, b, c = matrix[0]
        d, e, f = matrix[1]
        g, h, i = matrix[2]
        return a * (e*i - f*h) - b *(d*i - f*g) + c * (d*h - e*g) #(a(ei-fh)-b(di-fg)+c(dh-eg))

matrix = [[2, 3, 5],
          [4, 5, 8],
         [1, 5, 7]]

determinant_3x3(matrix)
```
